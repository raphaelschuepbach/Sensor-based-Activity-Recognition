# ML Model Evaluieren
Hier wird geschaut, welches Modell  der 6 ML Modellen ab besten abschneidet

```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math


plot_dir = Path(
    "..\\..\\Model_data\\ML_Daten\\PerformancePlots"
)

plot_files = sorted(plot_dir.glob("*.png"))

n = len(plot_files)
cols = 2
rows = math.ceil(n / cols)

fig, axes = plt.subplots(
    rows,
    cols,
    figsize=(16, 6 * rows)
)

axes = axes.flatten()

for ax, plot_path in zip(axes, plot_files):

    img = mpimg.imread(plot_path)

    ax.imshow(img)
    ax.axis("off")
    ax.set_title(plot_path.stem)

# leere Felder entfernen
for ax in axes[len(plot_files):]:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
